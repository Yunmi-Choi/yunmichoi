* 유저별로 특정 제품을 장바구니에 담거나 구매한 행동만을 보기 위해 groupby 데이터셋 생성
* 이때, 아이폰 유저는 장바구니 이벤트가 누실되었기에 안드로이드 유저 한정으로 진행

```python
import pandas as pd 
```

```python
df = pd.read_csv('C:/.../df_final.csv')
df 
```

```python
# 안드로이드만 필터링
df_a = df.loc[df['platform']=='ANDROID']
```

```python
# 장바구니와 구매 이벤트만 필터링
df_acp = df_a[(df_a['event_name']=='add_to_cart') | (df_a['event_name']=='purchase_success')]
```

```python
# 유저, 상품, 이벤트 기준으로 그룹바이
# 뒤의 ['item_name'].count()는 groupby()함수가 돌아가기 위해 아무거나 쓴 것이며 드랍할 예정
df_group = df_acp.groupby(['user_no', 'item_no', 'event_name'], as_index=False)['item_name'].count() 
df_group
```

```python
# 그룹바이한 기준 인덱스만 남기고 뒤의 집계는 불필요해 드랍
df_group.drop(columns=['item_name'], inplace=True)
df_group
```

```python
# 장바구니는 0, 구매는 1인 함수 제작
def unique_rank(x):
    if x == 'add_to_cart':
        return '0'
    else:
        return '1'    
```

```python
# 구매여부 컬럼 추가
df_group['purchase_bool'] = df_group['event_name'].apply(unique_rank)
```
